### 0b, Quantity of Sales per Year - Line Chart Exploration

```python
df.groupby("Year")["Quantity"].sum().sort_index(ascending=True)
```

```python
df.describe().T
```

```python
import duckdb

df_quantity_sold = duckdb.query(
    """
    SELECT
        Year,
        SUM(Quantity) as Quantity_sold,
    FROM df
    GROUP BY
        Year
    ORDER BY
        Year ASC
    """
).df()

df_quantity_sold
```

```python
df_quantity_sold = df_quantity_sold[(df_quantity_sold["Year"] >=2009) & (df_quantity_sold["Year"]!= 2017)]
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

fig, ax = plt.subplots(1, figsize= (12, 6), dpi=150)

ax.plot(df_quantity_sold["Year"], df_quantity_sold["Quantity_sold"], marker='o')
#format yaxis with 'K' for x 1000
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f"{int(x/1000)} K"))
ax.set(xlabel = "Year", ylabel= "Quantity", title = "Cars Sold Yearly")

plt.show()
```

```python
fig.savefig("car_sales_2009_to_2016.png")
```

### 0 c, Line chart of Avg. C02 Emissions for Same Years as In b,

```python
df.head()
```

```python
import duckdb

df_CO2_emissions = duckdb.query(
    """
    SELECT
        Year,
        SUM(Avg_CO2) as Avg_CO2_Emissions,
    FROM df
    GROUP BY
        Year
    ORDER BY
        Year ASC
    """
).df()

df_CO2_emissions
```

```python
df_CO2_emissions = df_CO2_emissions[(df_CO2_emissions["Year"] >=2009) & (df_CO2_emissions["Year"]!= 2017)]
```

```python
fig, ax = plt.subplots(1, figsize= (12, 6), dpi=150)

ax.plot(df_CO2_emissions["Year"], df_CO2_emissions["Avg_CO2_Emissions"], marker='o')

ax.set(xlabel = "Year", ylabel= "KtCO2e", title = "Yearly CO2 Emissions")

plt.show()
```
