### 0b, Quantity of Sales per Year - Line Chart Exploration

```python
df.groupby("Year")["Quantity"].sum().sort_index(ascending=True)
```

```python
df.describe().T
```

```python
import duckdb

df_quantity_sold = duckdb.query(
    """
    SELECT
        Year,
        SUM(Quantity) as Quantity_sold,
    FROM df
    GROUP BY
        Year
    ORDER BY
        Year ASC
    """
).df()

df_quantity_sold
```

```python
df_quantity_sold = df_quantity_sold[(df_quantity_sold["Year"] >=2009) & (df_quantity_sold["Year"]!= 2017)]
```

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick

fig, ax = plt.subplots(1, figsize= (12, 6), dpi=150)

ax.plot(df_quantity_sold["Year"], df_quantity_sold["Quantity_sold"], marker='o')
#format yaxis with 'K' for x 1000
ax.yaxis.set_major_formatter(mtick.FuncFormatter(lambda x, _: f"{int(x/1000)} K"))
ax.set(xlabel = "Year", ylabel= "Quantity", title = "Cars Sold Yearly")

plt.savefig("cars_sold_2009_to_2016.png", dpi=300, bbox_inches="tight")
plt.show()
```

#### Basically Pandas dataframe index is the order for plotting. So draw plot using that

```python
fig, ax = plt.subplots(figsize=(12,6), dpi=150)

ax.plot(df.index, df["Import"], marker="o") # chart follows index ascending
#add year boundaries from the DF
years = df["Year"].unique() #tracks each year
year_positions = df.groupby("Year").head(1).index #index position of first row per new year

ax.set_xticks(year_positions) #where on index to place labels
ax.set_xticklabels(years) #label for 'Year'

ax.set(xlabel="Month and Year", ylabel="Import Numbers", title="Import Numbers 2007-2017")

plt.savefig("imports_2007_to_2017.png", dpi=300, bbox_inches="tight")
plt.show()
```

### 0 f, Line Chart Tracking Electric Car Imports (in Months + Years where relevant)

```python
df.describe().T
```

```python
df[["Import_Electric", "Year"]].groupby(["Year"]).mean(numeric_only=True).sort_values(by="Year", ascending=True).reset_index()
```

#### Import Numbers Start in 2012, make DF tracking only 2012-2017

```python
# Remove years before 2012
df_import_electric = df[df["Year"] >=2012][["Year", "Month", "Import_Electric"]]

df_import_electric.head(12)
```

```python
#slice df for plotting to start in September 2012
df_import_electric = df_import_electric[df_import_electric["Year"] >= 2012]
```

#### Plot Electric Imports 2012 -2017

```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(12,6), dpi=150)

ax.plot(df_import_electric.index, df_import_electric["Import_Electric"], marker="o") # chart follows index ascending
#add year boundaries from the DF
years = df_import_electric["Year"].unique() #tracks each year
year_positions = df_import_electric.groupby("Year").head(1).index #index position of first row per new year

ax.set_xticks(year_positions) #where on index to place labels
ax.set_xticklabels(years)
ax.set_xlim(df_import_electric.index[0], df_import_electric.index[-1])

ax.set(xlabel="Month and Year", ylabel="Imports x 1000", title="Electric Car Imports 2012-2017")
#annotate first data point as it starts in the middle of the year
first_idx = df_import_electric["Import_Electric"].first_valid_index()
first_val = df_import_electric.loc[first_idx, "Import_Electric"]

ax.annotate(
            "September 2012",
            xy=(first_idx, first_val),
            xytext=(0, 50),
            textcoords="offset points",
            ha="center", color="green",
            fontsize=12, fontweight="bold",
            zorder=5,
            annotation_clip=False,
            arrowprops=dict(arrowstyle="-", color="green", lw=1.5)
            )

plt.savefig("Electric_Car_Imports_2012_to_2017.png", dpi=300, bbox_inches="tight")
plt.show()
```
