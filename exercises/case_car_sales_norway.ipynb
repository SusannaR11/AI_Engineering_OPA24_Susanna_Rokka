```python
import duckdb

df_quantity_sold = duckdb.query(
    """
    SELECT
        Year,
        SUM(Quantity) as Quantity_sold,
    FROM df
    GROUP BY
        Year
    ORDER BY
        Year ASC
    """
).df()

df_quantity_sold
```

```python
df_quantity_sold = df_quantity_sold[(df_quantity_sold["Year"] >=2009) & (df_quantity_sold["Year"]!= 2017)]
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, figsize= (12, 6), dpi=150)

ax.plot(df_quantity_sold["Year"], df_quantity_sold["Quantity_sold"], marker='o')
ax.set(xlabel = "Year", ylabel= "Quantity", title = "Cars Sold Yearly")

plt.show()
```

#### Based on plot years >= 2009 and 2017 should be dropped as they throw out the curve, 2017 may not be a complete data set for the entire year. 2008 saw a huge dip from 2007.
