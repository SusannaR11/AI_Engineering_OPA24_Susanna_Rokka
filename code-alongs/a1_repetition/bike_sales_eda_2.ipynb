## Profit per Age Group

```python
import duckdb

df_small_groups = duckdb.query(
    """
    SELECT
        age_group,
        customer_gender,
        country,
        SUM(profit) AS Total_Profit,
        COUNT(*) AS Number_Rows 
    FROM
        df
    GROUP BY
        age_group, customer_gender, country
    HAVING      --note difference btw 'having' and 'where'
        number_rows < 1000
    ORDER BY
        Total_Profit DESC
    """
).df()

df_small_groups
```

```python
df_senior_female = df_small_groups.query("Age_Group == 'Seniors (64+)' and Customer_Gender == 'F' ")
df_senior_female # query function vs SQL above. 'No for-loops' in Pandas, loses the power of Pandas, as Python is slow
```

```python
df_senior_female.plot(kind = "bar", x= "Country", y="Total_Profit")
```

```python
ax =df_senior_female.plot(kind = "barh", x= "Country", y="Total_Profit")
ax.set(title= "64+ females total profit in bike store", 
       xlabel="Total profit", 
       ylabel="Country",)
ax.invert_yaxis()
fig = ax.get_figure()
fig.tight_layout()
fig.savefig("profit_per_country_senior_females.png", dpi=200)
```
